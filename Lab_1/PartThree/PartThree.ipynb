```python
import pandas as pd
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix, classification_report
import matplotlib.pyplot as plt
```

```python
cardio_complete_data_set = pd.read_csv("cardio-complete.csv")
cardio_complete_data_set_data_poly = cardio_complete_data_set.iloc[:, 1:].copy()
```

```python
#One Hot Encode the Three columns 
cardio_complete_data_set_data_poly.gender = cardio_complete_data_set_data_poly.gender.map({'Men':0, 'Women':1})
cardio_complete_data_set_data_poly.gluc = cardio_complete_data_set_data_poly.gluc.map({'Normal':1, 'Above Normal':2, "High":3})
cardio_complete_data_set_data_poly.cholesterol = cardio_complete_data_set_data_poly.cholesterol.map({'Normal':1, 'Above Normal':2, "High":3})
```

```python
X_train, X_test, y_train, y_test = train_test_split(cardio_complete_data_set_data_poly.iloc[:, :-1], cardio_complete_data_set_data_poly.iloc[:, -1])
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
polynomial_logistic_regression_model = LogisticRegression(max_iter=1000)
accuracies = []
for i in range(1, 4):
    polynomial_transform = PolynomialFeatures(degree = i)
    X_train = polynomial_transform.fit_transform(X_train)
    X_test = polynomial_transform.fit_transform(X_test)

    
    polynomial_logistic_regression_model = LogisticRegression(max_iter=1000)
    polynomial_logistic_regression_model.fit(X_train, y_train)
    predictions_polynomial = polynomial_logistic_regression_model.predict(X_test)

    print(accuracy_score(y_test, predictions_polynomial),
          mean_squared_error(y_test, predictions_polynomial, squared=False),
          confusion_matrix(y_test, predictions_polynomial),
          sep='\n')
    accuracies.append(accuracy_score(y_test, predictions_polynomial))
```

```python
plt.subplots(1)
plt.plot(accuracies)
plt.title("Accuracies vs. Degree count")
```

```python
# Now we determine if model is under or overfitting.
model_kfold = LogisticRegression(max_iter=1000)

training_sizes = [x for x in range(10, 901, 10)]
features = ['age', 'weight', 'height', 'ap_lo', 'ap_hi', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'gender']
target = ['cardio']
```

```python
train_sizes, train_scores, validation_scores = learning_curve(
    estimator=model_kfold, X=cardio_complete_data_set_data_poly[features],
    y=cardio_complete_data_set_data_poly[target].values.ravel(),
    train_sizes=training_sizes,
    cv=10,
    scoring='neg_mean_squared_error'
)

train_scores_mean = -train_scores.mean(axis=1)
validation_scores_mean = -validation_scores.mean(axis=1)
print(pd.Series(train_scores_mean, index=training_sizes))
print(pd.Series(validation_scores_mean, index=training_sizes))
```

```python
plt.style.use('seaborn')
plt.plot(training_sizes, train_scores_mean, label = "Training Error")
plt.plot(training_sizes, validation_scores_mean, label = "Validation Error")
plt.ylabel("MSE", fontsize=10)
plt.xlabel("Training sizes")
plt.title("Learning curves for a logistic regression model")
plt.legend()
```
