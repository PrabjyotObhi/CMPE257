### Load sklearn's iris dataset and perform classification using Random Forest. Also plot the feature importances. Write down your comments.

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
iris = datasets.load_iris()

print("Total Number of Observations:", len(iris.target))
print("Feature Names:", iris.feature_names)
print("Target Names:", iris.target_names)
```

```python
X_iris = iris.data
y_iris = iris.target

X_train, X_test, y_train, y_test_iris = train_test_split(X_iris, y_iris, test_size = 0.30, random_state = 1)
```

```python
random_forest = RandomForestClassifier()
```

```python
rfcMaxDepthNone = RandomForestClassifier(max_depth= None)
rfcMaxDepthThree = RandomForestClassifier(max_depth= 3)
rfcMaxDepthFive = RandomForestClassifier(max_depth= 5)
rfcMaxDepthSeven = RandomForestClassifier(max_depth= 7)

rfcMaxDepthNoneEntropy = RandomForestClassifier(max_depth= None, criterion = 'entropy')
rfcMaxDepthThreeEntropy = RandomForestClassifier(max_depth= 3, criterion = 'entropy')
rfcMaxDepthFiveEntropy = RandomForestClassifier(max_depth= 5, criterion = 'entropy')
rfcMaxDepthSevenEntropy = RandomForestClassifier(max_depth= 7, criterion = 'entropy')
```

```python
rfcMaxDepthNone.fit(X_train,y_train)
rfcMaxDepthThree.fit(X_train,y_train)
rfcMaxDepthFive.fit(X_train,y_train)
rfcMaxDepthSeven.fit(X_train,y_train)


rfcMaxDepthNoneEntropy.fit(X_train,y_train)
rfcMaxDepthThreeEntropy.fit(X_train,y_train)
rfcMaxDepthFiveEntropy.fit(X_train,y_train)
rfcMaxDepthSevenEntropy.fit(X_train,y_train)


print('Models trained')
```

```python
# Now predicting for the test set.
y_predictions_rfcMaxDepthNone = rfcMaxDepthNone.predict(X_test)
y_predictions_rfcMaxDepthThree = rfcMaxDepthThree.predict(X_test)
y_predictions_rfcMaxDepthFive = rfcMaxDepthFive.predict(X_test)
y_predictions_rfcMaxDepthSeven = rfcMaxDepthSeven.predict(X_test)

y_predictions_rfcMaxDepthNoneEntropy = rfcMaxDepthNoneEntropy.predict(X_test)
y_predictions_rfcMaxDepthThreeEntropy = rfcMaxDepthThreeEntropy.predict(X_test)
y_predictions_rfcMaxDepthFiveEntropy = rfcMaxDepthFiveEntropy.predict(X_test)
y_predictions_rfcMaxDepthSevenEntropy = rfcMaxDepthSevenEntropy.predict(X_test)


# We can see that the results are stored in the form of array.
print(y_predictions_rfcMaxDepthNone)
print(y_predictions_rfcMaxDepthThree)
print(y_predictions_rfcMaxDepthFive)
print(y_predictions_rfcMaxDepthSeven)


print(y_predictions_rfcMaxDepthNoneEntropy)
print(y_predictions_rfcMaxDepthThreeEntropy)
print(y_predictions_rfcMaxDepthFiveEntropy)
print(y_predictions_rfcMaxDepthSevenEntropy)
```

```python
print('Accuracy for given y_predictions_MaxDepthNone split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthNone))
print('Accuracy for given y_predictions_MaxDepthThree split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthThree))
print('Accuracy for given y_predictions_MaxDepthFive split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthFive))
print('Accuracy for given y_predictions_MaxDepthSeven split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthSeven))


print('\n')
print('Accuracy for given y_predictions_MaxDepthNoneCriterionEntropy split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthNoneEntropy))
print('Accuracy for given y_predictions_MaxDepthThreeCriterionEntropy split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthThreeEntropy))
print('Accuracy for given y_predictions_MaxDepthFiveCriterionEntropy split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthFiveEntropy))
print('Accuracy for given y_predictions_MaxDepthSevenCriterionEntropy split = ',accuracy_score(y_test_iris, y_predictions_rfcMaxDepthSevenEntropy))

#All of the accuracies are the same, so i'll only print one set feature importance
```

## Write four differences between Random Forests and Decision Trees 


1. A decision tree is only one tree while a random forest is multiple trees
2. Decision trees utilize all features to make its decisions whereas each tree in a random forest is a little bit different, and they'd vote on what the outcome would be
3. Decision trees are much easier to interpret and see whereas a random forest is more like a blackbox -- you dont really know what the trees themselves look like but you know how many you would have set using the n_estimators
4. Decision Trees are easier to overfit -- no maxdepth
